```python
import random
import numpy as np
import matplotlib.pyplot as plt

X_Coordinates = []
Y_Coordinates = []
x_inner_circle = []
y_inner_circle = []
x_outter_circle = []
y_outter_circle = []
pi_values = []
radius = 0.5


def Points_In_Circle(num_of_points):
  points_in_circle = 0
  i=0
  X_Coordinates = np.array([random.uniform(0,1) for _ in range(num_of_points)])
  Y_Coordinates = np.array([random.uniform(0,1) for _ in range(num_of_points)])
  for i in range(num_of_points):
    distance = (X_Coordinates[i]-0.5)**2 + (Y_Coordinates[i]-0.5)**2
    if distance <= radius*radius:
        points_in_circle += 1
        x_inner_circle.append(X_Coordinates[i])
        y_inner_circle.append(Y_Coordinates[i])
    else:
        x_outter_circle.append(X_Coordinates[i])
        y_outter_circle.append(Y_Coordinates[i])
  return points_in_circle,x_inner_circle,y_inner_circle,x_outter_circle,y_outter_circle
```

```python

def plot(Xin,Yin,Xout,Yout):
  circle = plt.Circle((0.5, 0.5), radius, color='blue',fill=False)
  fig, ax = plt.subplots(figsize=(5,5))
  plt.xlim(0,1)
  plt.ylim(0,1)
  ax.add_artist(circle)
  plt.scatter(x_inner_circle,y_inner_circle,color="black")
  plt.scatter(x_outter_circle,y_outter_circle,color="red")
  plt.show()
```

```python
pi_values = []
j=0
list_of_num_of_points = [10,500,1000,5000,10000,50000,80000,100000,200000]
for j in range(len(list_of_num_of_points)):
  points,xinner,yinner,xoutter,youtter = Points_In_Circle(list_of_num_of_points[j])
  pivalue = 4 * points / list_of_num_of_points[j]
  plot(xinner,yinner,xoutter,youtter)
  pi_values.append(pivalue)

print('\n\nPi values in given list:',pi_values)
```
